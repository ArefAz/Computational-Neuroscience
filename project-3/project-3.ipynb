In this project, we implement two types of connections, $DenseConnection$ and
$RandomConnection$. The $DenseConnection$'s weights are drawn from a normal
distribution with $\mu=j_0/N$ and $\sigma=\sigma_0/N$. If a drawn weight
from this distribution lie outside of $[w_{min}, w_{max}]$, it will be set to
$\mu=j_0/N$.

Also, the scheme used for $RandomConnection$
is **fixed number of pre-synaptic partners**, so we don't need to scale the weights.
$w_{max}=1$, and $w_{min}=0$ is fixed for all the experiments.
The weights in $RandomConnection$ are set to $\frac{(w_{max} - w_{min})}{2}$, which by
default is equal to $0.5$. This value is fixed for all the experiments.



We use three connections:

1. Excitatory population to itself

2. Excitatory population to inhibitory population

3. Inhibitory population to excitatory population

In all the experiments, the types of connections will be equal among the three
connections, i.e., if one of the connections is dense, the other two are dense
as well.

In each experiment, only one parameter will be changed compared to the previous
parameter set to emphasise the effect of that specific parameter.

The input will be a pulse current with a mean value of $20$, and a low amount of noise. Also, the inputs
to the neurons (in one population) slightly differ from one another with a
standard deviation of $0.25$. This inter-neuron input noise is added to have some
variation in each neuron's response. The time-noise standard deviation is set to $1$.

For comparing the total activity of the populations during the simulation time,
we calculate the area under the curve $(AUC)$ of activity-plot. Higher $AUC$ indicates
higher total activation.



```python
import sys
sys.path.append('..')

from cnsproject.plotting.plotting import draw_connection_plots, input_plot
from cnsproject.network.connections import DenseConnection, RandomConnection
from cnsproject.utils import run_connection_simulation

%matplotlib inline
```

```python
# Just to draw the input
m1, _ = run_connection_simulation(connection_type=DenseConnection)
input_plot(m1, figsize=(15, 7))
```

 the input each neuron receives is slightly different.

# Dense Connection

The parameters of this connection is: $J_0$ and $\sigma_0$ for each connection,
and the total number of neurons. $J_0$ directly controls the weights of the
synapses and therefore the strength of that synapse. $\sigma_0$ controls the
amount of variation in the synapses' weights.


## Parameter set 1:
$J_{EE}=16$, $\sigma_{EE}=0$,

$J_{EI}=10$, $\sigma_{EI}=0$,

$J_{IE}=10$, $\sigma_{IE}=0$,

$N=100$


```python
j0 = [16, 10, 10]
sig0 = [1e-10, 1e-10, 1e-10]
n = 100
m1, m2 = run_connection_simulation(
    connection_type=DenseConnection,
    j0=j0, sig0=sig0, n_neurons=n
)
title = "$J_0 = {}$, $\sigma_0 = {}$, $N = {}$".format(j0, [0, 0, 0], n)
draw_connection_plots(m1, m2, title=title)
```

We see that both populations' AUC is equal, and by observing the plots we see that
both of them, are showing activation in the same time, this means that we have
a balanced connection between the two population.

## Parameter set 4:

Testing $J_{EE}$'s effect

$J_{EE}=22$, $\sigma_{EE}=0$,

$J_{EI}=10$, $\sigma_{EI}=0$,

$J_{IE}=10$, $\sigma_{IE}=0$,

$N=100$

```python
j0 = [22, 10, 10]
sig0 = [1e-10, 1e-10, 1e-10]
n = 100
m1, m2 = run_connection_simulation(
    connection_type=DenseConnection,
    j0=j0, sig0=sig0, n_neurons=n,
)
title = "$J_0 = {}$, $\sigma_0 = {}$, $N = {}$".format(j0, [0, 0, 0], n)
draw_connection_plots(m1, m2, title=title)
```

By increasing $J_{EE}$ (increasing the weights of the synapses in population E,
in fact), the total activation of the populations will increase,
since the excitatory populations' spikes will excite itself, and its activation
will increase. Also, the increase in activation of the first population will
excite the second population, so the second population's activity will
increase as well.

## Parameter set 5:

Testing $J_{EI}$'s effect

$J_{EE}=10$, $\sigma_{EE}=0$,

$J_{EI}=20$, $\sigma_{EI}=0$,

$J_{IE}=10$, $\sigma_{IE}=0$,

$N=100$

```python
j0 = [10, 20, 10]
sig0 = [1e-10, 1e-10, 1e-10]
n = 100
m1, m2 = run_connection_simulation(
    connection_type=DenseConnection,
    j0=j0, sig0=sig0, n_neurons=n,
)
title = "$J_0 = {}$, $\sigma_0 = {}$, $N = {}$".format(j0, [0, 0, 0], n)
draw_connection_plots(m1, m2, title=title)
```

Increasing the value of $J_{EI}$ has an interesting effect; population I's activty
will not increase, but the activation of population E will decrease. This observation
could be explained as follows:
By increasing $J_{EI}$, the population I will be more effected from
population E's activity. This, eventually, reduce the activity of population E,
but not much to significantly reduce the population I's activity. If we carefully
observe the plot, we see that E's activity is vanished, and I's activity is decreased
after the first interval of activation. The decrease in I's activity then let the E to be active
in the next time steps. After that, they reach to a state of balance, and the cycles will go on.
We could easily see that how the second population is
controlling the activity of the first population.

# Random Connection

Since the random schema we implemented is the **fixed number of pre-synaptic partners**,
the only parameters we have is number of pre-synaptic connections for each post-synaptic
neuron. The weights for all the synapses are equal to $0.5$ as explained in the
top of this report. We also explore the effect of neurons' count.

## Parameter set 1:

Testing the effect of $n_{EE}$

$n_{EE}: [0, 60]$

$n_{EI}=20$

$n_{IE}=10$

$N=100$

Note: $n_{EI}=20$ means that each neuron in the population I,
has 20 randomly sampled pre-synaptic connections to the population E from
the set of all possible connections to population E.


```python
n = 100
for i in range(0, 60, 10):
    num_connections = [i, 15, 10]
    m1, m2 = run_connection_simulation(
        connection_type=RandomConnection,
        num_connections=num_connections, n_neurons=n
    )
    title = "$n_{{connections}} = {}$, $N = {}$".format(num_connections, n)
    draw_connection_plots(m1, m2, title=title)
```

If $n_{EE}$ is too low, the inhibitory population will prevent the normal activity
of population E. By increasing it, we could achieve a balanced network where
two populations work the same activity. If $n_{EE}$ gets too high, population E
activation will be saturated and population I cannot prevent this no matter
how high its own activation is.


## Parameter set 2:
Testing the effect of $n_{EI}$

$n_{EE}=30$

$n_{EI}: [0, 25]$

$n_{IE}= 10$

$N=100$

```python
n = 100
for i in range(0, 30, 5):
    num_connections = [30, i, 10]
    m1, m2 = run_connection_simulation(
        connection_type=RandomConnection,
        num_connections=num_connections, n_neurons=n
    )
    title = "$n_{{connections}} = {}$, $N = {}$".format(num_connections, n)
    draw_connection_plots(m1, m2, title=title)
```

From the above plots in which we only changed $n_{IE}$ incrementally, the effect
of $n_{IE}$ is visible clearly. When $n_{IE}$ is low, the activation of
population E is more than the population I's activity. By increasing $n_{IE}$,
we can reach a state the activations are equal. When $n_{IE}$ becomes too high,
the activation of population E tends to deteriorate.

## Parameter set 3:

Testing the effect of $n_{IE}$

$n_{EE}=30$

$n_{EI}=15$

$n_{IE}: [2, 20]$

$N=100$

```python
n = 100
for i in range(0, 20, 5):
    num_connections = [30, 15, i]
    m1, m2 = run_connection_simulation(
        connection_type=RandomConnection,
        num_connections=num_connections, n_neurons=n
    )
    title = "$n_{{connections}} = {}$, $N = {}$".format(num_connections, n)
    draw_connection_plots(m1, m2, title=title)
```

We see that the effect of $n_{IE}$ and $n_{EI}$ is almost the same.
The conclusion is that by tuning the number of connections in the random scheme,
we could have a balanced network,
which acts very similar to the balanced network we created using dense connections.
